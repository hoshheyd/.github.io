```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 1. Definiera den Symplektiska cylindern (Begränsningen)
theta = np.linspace(0, 2*np.pi, 50)
z = np.linspace(-3, 3, 50)
Theta, Z = np.meshgrid(theta, z)
R_cyl = 1.0 
X_cyl = R_cyl * np.cos(Theta)
Y_cyl = R_cyl * np.sin(Theta)

# 2. Definiera fasrumsballongen (Det kollapsande SAGD-kärnan)
u = np.linspace(0, 2 * np.pi, 50)
v = np.linspace(0, np.pi, 50)
a, b, c = 0.95, 0.95, 2.8 
X_ell = a * np.outer(np.cos(u), np.sin(v))
Y_ell = b * np.outer(np.sin(u), np.sin(v))
Z_ell = c * np.outer(np.ones(np.size(u)), np.cos(v))

# 3. Visualisering
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X_cyl, Y_cyl, Z, alpha=0.1, color='cyan')
ax.plot_surface(X_ell, Y_ell, Z_ell, alpha=0.8, color='red')
ax.set_xlabel('q1'), ax.set_ylabel('q2'), ax.set_zlabel('p1 (Torsion)')
plt.title('SAGD Phase-Space: Topological Non-Squeezing')
plt.show()
```
